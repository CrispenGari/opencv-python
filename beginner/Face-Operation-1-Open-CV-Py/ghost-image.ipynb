# Face Operation Project
* This is a simple `open-cv` face operation that creates a ghost image from another image

```python
### Imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Read an image and display It
* The image we are going to read is in the `images` directory with name 'me.jpg'

```python
image = cv2.imread("images/me.jpg")
imageRgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show
plt.imshow(imageRgb)
plt.show()
```

## Use Image pyramind
* [Image Pyraminds](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_pyramids/py_pyramids.html)

```python
imageSmall = cv2.pyrDown(imageRgb)
plt.imshow(imageSmall)
plt.show()
```

```python
for i in range(5):
    imageSmall= cv2.bilateralFilter(imageSmall, d=9, sigmaColor=9, sigmaSpace=7)

imageRgb = cv2.pyrDown(imageRgb)
plt.imshow(imageRgb)
```

## Convert the original image to gray scale


```python
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(imageGray, cmap="gray")
plt.show()
```

```python
## Apply the medianBlur on an image
```

```python
imageBlur = cv2.medianBlur(imageGray, 1)
plt.imshow(imageBlur, cmap="gray")
plt.show()
```

## Apply adaptiveThreshold to the image to reduce some noise

```python
imageEdge = cv2.adaptiveThreshold(imageBlur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7, 2)
```

```python
plt.imshow(imageEdge)
```

```python
imageEdge = cv2.cvtColor(imageEdge, cv2.COLOR_GRAY2RGB)
plt.imshow(imageEdge)
```

```python
ghost = cv2.bitwise_and(image, imageEdge)
```

```python
plt.imshow(ghost)
```
